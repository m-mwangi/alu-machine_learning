```python
"""
Histogram that depicts the distribution of carbon Dioxide Emissions
Sets the number of bins to 20, x axis being carbon dioxide emission
while y axis showing the frequency
"""

co2_values=df['co2']
sns.histplot(data=co2_values,bins=20)
plt.xlabel('Carbon Dioxide Emission')
plt.ylabel('Frequency')
plt.title('Distribution of carbon Dioxide among cars')
plt.savefig('co2_distribution')
plt.show()
```

**Interpretation**
Most cars emit high levels of CO2 shown from the heights of most bars being high.
The height of the bars which represent the count of vehicles.

```python
"""
Histogram that depicts the Engine displacement
Sets the number of bins to 10, x axis being engine displacement
while y axis showing the frequency
"""

engine_dis=df['displ']
sns.histplot(data=engine_dis,bins=10,color='green')
plt.xlabel('Engine Displacement (L)')
plt.ylabel('Frequency')
plt.title("Distribution of Engine Displacement")
plt.savefig('engine_displacement')
plt.show()
```

**Interpretation**
Most vehicles(over 1200) have an engine displayment of 2.0

```python
"""
Histogram that depicts Highway Fuel Efficiency
Sets the number of bins to 10, x axis being engine displacement
while y axis showing the frequency
"""

fuel_eff=df['highway']
sns.set(style="whitegrid")
sns.histplot(data=fuel_eff,bins=10,color='red')
plt.xlabel('Highway Fuel Efficiency')
plt.ylabel('Frequency')
plt.title("Distribution of Highway Fuel Efficiency")
plt.savefig('highway_fig')
plt.show()
```

**Interpretation**
 Few vehicles exhibit high highway fuel efficiency, as indicated by the low frequency of occurrences at the upper end of the fuel efficiency spectrum.

```python
num_values = df.select_dtypes(include=['float64', 'int64'])
corr_matrix = num_values.corr()
plt.figure(figsize=(9,6))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.show()
```

**Interpretation** The heat map shows correlation between variables. Those variables that strong positive correlation have a dark red colour while those with strong negative correlations coloured in dark blue and 0 means no correlation which lightly coloured.
For examples; there is a strong positive correlation between city and combined fuel efficiency.
