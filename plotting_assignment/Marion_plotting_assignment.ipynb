```python
#create a histogram that depicts the carbon dioxide distribution
#Set the number of bins to 5, with the x axis being CO2 emissions while y axis showing the frequency

co2_emissions = data['co2']
sns.histplot(data = co2_emissions,bins = 5, color = 'green')
plt.xlabel('CO2 Emissions')
plt.ylabel('Frequency')
plt.title('Carbon Dioxide Distribution Among Cars')
plt.savefig('co2_emission')
plt.show()
```

Graphical Intepretation- Most cars above 1200, have a CO2 emission of between 300 to 400.

```python
#create a histogram that depicts the engine displacement
#Set the number of bins to 20, with the x axis being engine displacement while y axis showing the frequency

engine_displacement = data['displ']
sns.histplot(data = engine_displacement,bins = 20, color = 'red')
plt.xlabel('Engine Displacement (L)')
plt.ylabel('Frequency')
plt.title('Distribution of Engine Displacement among cars')
plt.savefig('engine_displacement')
plt.show()
```

Graphical Interpretation- Majority of the vehicles above 800 have an engine displacement of 2.

```python
##create a histogram that depicts the highway fuel efficiency
#Set the number of bins to 10, with the x axis being highway fuel efficiency while y axis showing the frequency


fuel_eff = data['highway']
sns.set(style="whitegrid")
sns.histplot(data = fuel_eff,bins = 10,color = 'blue')
plt.xlabel('Highway Fuel Efficiency')
plt.ylabel('Frequency')
plt.title("Distribution of Highway Fuel Efficiency Among Cars")
plt.savefig('highway_fuel_eff')
plt.show()
```

```python
num_values = data.select_dtypes(include=['float64', 'int64'])
corr_matrix = num_values.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.show()
```

**Interpretation:**

The heatmap displays the correlation between variables. Variables with a strong positive correlation are shown in dark red, while those with a strong negative correlation appear in dark blue. A value of 0, indicating no correlation, is represented by light colors. For example, there is a strong positive correlation between city fuel efficiency and combined fuel efficiency.
